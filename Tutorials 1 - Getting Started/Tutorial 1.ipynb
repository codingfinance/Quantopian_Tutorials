First import all the libraries.

```python
from quantopian.research import prices, symbols
import pandas as pd
import matplotlib.pyplot as plt
```

```python
aapl_close = prices(assets=symbols('AAPL'),
                   start = "2013-1-1",
                   end = "2016-1-1")
```

```python
aapl_close.head()
```

```python
aapl_close.plot();
```

Calculating the Simple Moving Average (SMA)

```python
sma_20 = aapl_close.rolling(window = 20).mean()
sma_50 = aapl_close.rolling(window = 50).mean()
```

```python
df = pd.DataFrame({'AAPL':aapl_close,
                  'SMA_20':sma_20,
                  'SMA_50':sma_50})
```

```python
df.plot()
plt.title("Apple Stock price with the 20 & 50 Day Moving average")
```
